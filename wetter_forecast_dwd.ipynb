## DWD Deutscher Wetterdienst

Forecast Data f√ºr Station-ID Kiel Holtenau 10046

```python
import requests
import json
import pandas as pd
url = 'https://dwd.api.proxy.bund.dev/v30/stationOverviewExtended?stationIds=10046'
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    display(data)
else:
    print("Fehler beim Abrufen der Daten:", response.status_code)
```

```python
import pandas as pd
import datetime

# Extract forecast data
forecast = data['10046']['forecast1']

# Calculate timestamps based on start time and time step
start_time = datetime.datetime.fromtimestamp(forecast['start'] / 1000)
time_step = forecast['timeStep'] // 1000  # Convert to seconds

timestamps = [start_time + datetime.timedelta(seconds=i * time_step) for i in range(len(forecast['temperature']))]

# Ensure all data arrays are of the same length by padding with None
def pad_list(lst, target_len):
    return lst + [None] * (target_len - len(lst))

# Find the maximum length among the data lists
max_len = max(len(timestamps), len(forecast['temperature']), len(forecast.get('sunshine', [])), len(forecast.get('dewPoint2m', [])), len(forecast.get('humidity', [])))

# Pad each list to the maximum length
timestamps = pad_list(timestamps, max_len)
temperature = pad_list(forecast['temperatureStd'], max_len)
sunshine = pad_list(forecast.get('sunshine', []), max_len)
dew_point = pad_list(forecast.get('dewPoint2m', []), max_len)
humidity = pad_list(forecast.get('humidity', []), max_len)

# Build DataFrame
df = pd.DataFrame({
    'Timestamp': timestamps,
    'Temperature': temperature,
    'Sunshine': sunshine,
    'DewPoint2m': dew_point,
    'Humidity': humidity
})

# Display the DataFramedf
df.head()
```

Observation:
- Not all data columns are forecasted in the same time range on hourly basis
    - temperature is forecasted for 10 days
    - sunshine, humidity, etc. for 5 days
